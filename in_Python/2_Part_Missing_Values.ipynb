```python
#First of all, let's import necessary libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('1_Lending_Club_final_variables.csv')
```

Now we can state that those variables which are with small value count
are categorical datatypes
We will have to get dummies for them in the future \
For now we have to get only EDA (exploratory data analysis)

```python
#Here you can see missing values
df.isna().sum().sort_values(ascending=False).head(10)
```

```python
#As we could see from unique values, here we simply visualize it
#The number of Fully paid is much higher
plt.figure(dpi=200)
sns.countplot(df['loan_status'])
```

```python
#let's see frequency distribution of loan_amount
plt.figure(dpi=200)
sns.distplot(df['loan_amnt'], bins=12)
plt.xticks(ticks=np.arange(0,45000, 5000))
plt.show()
#as we can see, we have picks with step 5000
#our mode seems to be 10000, median and mean higher correspondigly
```

```python
#Let's check our assumption
print('Mode   is ', df['loan_amnt'].mode().iloc[0])
print('Median is ', np.median(df['loan_amnt']))
print('Mean   is ', round(np.mean(df['loan_amnt'])))
#yes, our assumtion is quite well proved
#We have to mention that we have outliers since our mean deviated from mode
```

```python
#let's see our box plot with outliers on top
plt.figure(dpi=200)
sns.boxplot(y=df['loan_amnt'])
plt.show()
#as we can see, yes, they are on top
#our next step will be to equalize outliers to the high edge of boxplot
#and this should be applied to each contionius variable
```
